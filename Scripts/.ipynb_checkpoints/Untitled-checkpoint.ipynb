```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# centro (x,y) das estacoes A, B, C, D, respectivamente
energyStationsX =[+1.1250e+0, +3.1000e+0, -1.8000e+0, -4.3500e+0]
energyStationsY =[-2.4750e+0, +2.6250e+0, +1.3250e+0, -2.7750e+0]

sizeStation = 1.0

poweStationName = ['A', 'B', 'C', 'D']
p = 0

color = 0
colorGraphs = ['m','b','g']

with open('NeedSurviveLevel.txt', 'r') as f:
    y = f.read()
    lines = y.split()
tam = len(lines)
i = 0
energyValue = []
drive = []

while i < (tam-1):
    energyValue.append(float(lines[i]))
    i += 1
    drive.append(0)

    
with open('NeedExploreLevel.txt', 'r') as f:
    y = f.read()
    lines = y.split()
tam = len(lines)
i = 0
curiosityValue = []

while i < (tam-1):
    curiosityValue.append(float(lines[i]))
    i += 1
    drive.append(1)
    
posIniAll = []
posFimAll = []

j = 4
stop = j + 1
while (j < stop):
	with open('GroundTruthPosition0'+str(j)+'.txt', 'r') as f:
		x = f.read()
		lines = x.split()
	tam = len(lines)
	posTup = []

	i = 0

	while i < (tam-1):
		x = float(lines[i])
		y = float(lines[i+1])
		pos = [x, y]#[-y, x]
		posTup.append(pos)
		i = i+2

	posX, posY = zip(*posTup)

	x_ini = posX[0]
	y_ini = posY[0]

	x_fim = posX[len(posX)-1]
	y_fim = posY[len(posY)-1]

	posIt = [posX[0], posY[0]]
	posItF = [posX[len(posX)-1], posY[len(posY)-1]]

	posIniAll.append(posIt)
	posFimAll.append(posItF)

	with open('exploringActionsNumber0'+str(j)+'.txt', 'r') as f:
		actions = f.read()
	actions = actions.replace("\n", "")

	#ax.scatter(posX, posY, label = actions + ' actions')

	j = j + 1

	numpy_array =[]
	for e in zip(energyValue,curiosityValue,posX,posY):
		numpy_array.append(np.asarray(e))
    

```

```python
print(numpy_array)
df = pd.DataFrame(numpy_array, columns=['energy', 'curiosity', 'x', 'y'])
df.head()
```

```python
posXI, posYI = zip(*posIniAll)
posXF, posYF = zip(*posFimAll)
fig, ax = plt.subplots(figsize=(9, 7))

ax.scatter(posXF,posYF,c='r',marker='D', s=100,  zorder=3, label = 'End')

ax.scatter(posXI[0],posYI[0],c='k',marker='v', s=100, zorder=3, label = 'Start Pose')
#ax.scatter(posXI[1],posYI[1],c='k',marker='^', s=100, zorder=3)

c = 0
title = ['Trajectory: Long-term decisions','Trajectory: Short-term decisions']
nameSave = ['Trajectory_Long-term decisions','Trajectory - Short-term decisions']

#ax.grid(True)
ax.legend(loc='best', fontsize = 'large', ncol = 3)
plt.xlabel("x", fontsize=18)
plt.ylabel("y", fontsize=18)
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.title(title[c],  fontsize=18)
plt.ylim(-4.8,4.8)
plt.xlim(-4.8,4.8)
#plt.savefig(nameSave[c] + '.png')
#plt.show()
```
